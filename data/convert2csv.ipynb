```python
import numpy as np
import pandas as pd
```

```python
apdvlr = np.load('./clusters.CCF_APDVLR.npy')
```

```python
meta_df = pd.read_csv('./metadata.csv')
```

```python
meta_df
```

```python
apdvlr.shape
```

```python
df = pd.DataFrame(columns=['ap', 'dv', 'lr', 'area', 
                           'color','ignore',
                           'dur','wave'])
```

```python
for i in np.arange(0, apdvlr.shape[0]):
    row = meta_df.iloc[i]
    df.loc[i] = [apdvlr[i,0], apdvlr[i,1], apdvlr[i,2], row['new_label'], 
                 row['CCFColor'], 1 - row['keep'],
                 row['duration'] <= 0.4, row['footprint'] <= 20]
```

```python
df
```

```python
df.to_csv('./output.csv', index=False)
```
